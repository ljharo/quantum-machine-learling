```python
nbqubits = 20
qc = qiskit.QuantumCircuit(nbqubits)

qc = qiskit.QuantumCircuit(nbqubits)
qc.h(0)
qc.h(18)


for i in range(nbqubits):
    if i+1 < nbqubits:
        qc.cx(i, i+1)
    

stateVec = qiskit_quantum_info.Statevector.from_instruction(qc)
probavec = stateVec.probabilities()

labels = [format(i, f"0{nbqubits}b")[::-1] for i in range(2**nbqubits)]

if nbqubits < 5:

    # print(stateVec.data)
    # print(probavec)

    
    plt.bar(range(2**nbqubits), probavec)
    plt.xticks(range(2**nbqubits),labels=labels)
else:
    for i, val in enumerate(probavec):
        if val > 0:
            print(f"{round(val, 2)} = {labels[i]}")

print(f"{stateVec.data.nbytes/1024/1024:.0f}MB")

qc.draw(output='mpl')
```
